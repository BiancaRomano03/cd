```python
#medidas de tendencia central 

#age

df= pd.DataFrame(df,columns=["age"])
df.dropna(inplace=True)

#media 
media=df["age"].mean()
print("la media es:", media)

#mediana
mediana=df["age"].median()
print("la mediana es:", mediana)

#moda
moda=df["age"].mode()
print("la moda es:", moda)
```

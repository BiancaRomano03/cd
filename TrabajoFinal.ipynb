```python
#medidas de tendencia central 

#feature age (edad)
df=pd.DataFrame(df, columns=["Age"])

##media
media=df["Age"].mean()
print("La media de edad es:", media)

##mediana
mediana=df["Age"].median()
print("La mediana de edad es:", mediana)

##moda
moda=df["Age"].mode()
print("La moda de edad es:", moda)
```

```python
#feature sleep duration (duracion del sueño)

df=pd.DataFrame(df, columns=["Sleep Duration"]).fillna(0).astype('int64')

##media
media=df['Sleep Duration'].mean()
print("La media de la suracion del sueño es:", media)

##mediana
mediana=df['Sleep Duration'].median()
print("La mediana de la duracion del sueño es:", mediana)

##moda
moda=df['Sleep Duration'].mode()
print("La moda de la duracion del sueño es:", moda)
```

```python
#feature quality of sleep (calidad del sueño) 

df= pd.DataFrame(df,columns=["Quality of Sleep"]).fillna(0).astype('int64')

##media
media=df['Quality of Sleep'].mean()
print("La media de la calidad del sueño es:", media)

##mediana
mediana=df['Quality of Sleep'].median()
print("La mediana de la calidad del sueño es:", mediana)

##moda
moda=df['Quality of Sleep'].mode()
print("La moda de la calidad del sueño es:", moda)
```

```python
#feature physical activitiy level (nivel de actividad fisica) 

df=pd.DataFrame(df, columns=["Physical Activity Level"]).fillna(0).astype('int64')

##media
media=df["Physical Activity Level"].mean()
print("La media del nivel de actividad fisica:", media)

##mediana
mediana=df["Physical Activity Level"].median()
print("La mediana del nivel de actividad fisica es:", mediana)

##moda
moda=df["Physical Activity Level"].mode()
print("La moda del nivel de actividad fisicaes:", moda)
```
